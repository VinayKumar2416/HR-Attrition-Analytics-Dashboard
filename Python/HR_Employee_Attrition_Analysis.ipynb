# 6. Exploratory Data Analysis (EDA)

### What is the overall employee attrition rate in the organization?

```python
# Employee Attrition Rate

attrition_rate = round( df["Attrition"].value_counts(normalize=True) * 100,2 )

attrition_rate
```

```python
# Employee Count

df["Attrition"].value_counts()
```

```python
# Employee Attrition Rate Graph

df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(4,4),
    ylabel='',
    title="Employee Attrition Rate")

plt.show()
```

### Which department experiences the highest employee attrition?

```python
# Department-wise Attrition Count

department_attrition = pd.crosstab(df["Department"],df["Attrition"])

department_attrition
```

```python
department_attrition.plot( kind ='bar', stacked = True, figsize = (8,5))
plt.title("Department-wise Employee Attrition")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.xticks(rotation=0)
plt.show()
```

### Which job roles have the highest employee attrition?

```python
# Job Role-wise Attrition Count

jobrole_attrition = pd.crosstab( df["JobRole"], df["Attrition"])

jobrole_attrition
```

```python
# Job Role-wise Employee Attrition

jobrole_attrition.plot( kind ='barh', stacked = True, figsize = (10,6))
plt.title('Job Role-wise Employee Attrition')
plt.xlabel('Number of Employees')
plt.ylabel('Job Role')
plt.show()
```

### Does working overtime increase employee attrition?

```python
# Overtime vs Employee Attrition

overtime_attrition = pd.crosstab( df["OverTime"], df["Attrition"])

overtime_attrition
```

```python
overtime_attrition.plot( kind = "bar", stacked = True, figsize = (7,5))
plt.title("Overtime vs Employee Attrition")
plt.xlabel("Overtime")
plt.ylabel("Number of Employees")
plt.xticks(rotation=0)
plt.show()
```

### Which age groups experience the highest employee attrition?

```python
# Create Age Groups

age_bins = [18, 25, 35, 45, 55, 65]

age_labels = ['18-25','26-35', '36-45', '46-55', '56-65']

df['AgeGroup'] = pd.cut( df['Age'], bins = age_bins, labels = age_labels)
```

```python
age_attrition = pd.crosstab(df["AgeGroup"],df["Attrition"])
age_attrition
```

```python
age_attrition.plot( kind ='bar', stacked = True, figsize = (5,3))
plt.title("Age Group-wise Employee Attrition")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.xticks(rotation=0)
plt.show()
```

### Does monthly income influence employee attrition?

```python
# Monthly Income vs Employee Attrition

plt.figure(figsize=(8,5))

sns.boxplot( data = df, x ='Attrition', y ='MonthlyIncome')
plt.title('Monthly Income vs Employee Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

#### ## Key Business Insights

After performing exploratory data analysis on the HR Employee Attrition dataset, the following key insights were identified:

1. The overall employee attrition rate is **16.12%**, indicating that approximately one out of every six employees left the organization.

2. The Research & Development department recorded the highest number of employee exits due to its larger workforce.

3. Sales Executive, Research Scientist, and Laboratory Technician were the job roles with the highest employee attrition.

4. Employees who worked overtime exhibited significantly higher attrition than employees who did not work overtime.

5. Employee attrition was more common among younger employees, particularly those in the early stages of their careers.

6. Employees who left the organization generally had lower monthly income than employees who remained.

7. Lower job satisfaction appeared to be associated with higher employee attrition.
